```python
res = pickle.load(open("figs/rbm_learn/cyc_dmala_SbC_big_0.2_small_0.2_ss_10_total_20000.pickle", "rb"))
res = res[1234567]['burnin_res']
alpha_max = []
a_s_alphamax = []

for burnin_segment in res: 
    max_acc = np.argmax(burnin_segment["a_s"])
    alpha_max.append(burnin_segment["alpha_max"][max_acc])
    a_s_alphamax.append(burnin_segment["a_s"][max_acc])

    
    
```

```python
fig, ax = plt.subplots(2, figsize=(7, 10))
x = [i for i in range(len(alpha_max))]
ax[0].plot(x, alpha_max)
ax[0].set_ylabel("Alpha Max Step Size")
ax[0].grid()
ax[1].plot(x, a_s_alphamax)
ax[1].set_ylabel("Alpha Max A_s")
ax[1].grid()
plt.xlabel("Iterations (1,000s)")
```
