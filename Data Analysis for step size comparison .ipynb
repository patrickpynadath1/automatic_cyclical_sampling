```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
sns.set(color_codes=True)
sns.set_style("whitegrid")
from matplotlib import rcParams
rcParams.update({'figure.autolayout': True})
```

```python
cyc_dula_data = {}
dula_data = {}
step_size = [.2, .4, .5, 1.0]

for st_s in step_size:

    cyc_dula = np.load(f'./figs/ising_learn/rmse_cyc_dula_0.25_100_{st_s}.npy')
    dula = np.load(f'./figs/ising_learn/rmse_dula_0.25_100_{st_s}.npy')
    
    dula = [np.log(t) for t in dula]
    cyc_dula = [np.log(t) for t in cyc_dula]
    
    cyc_dula_data[st_s] = cyc_dula
    dula_data[st_s] = dula
```

```python
x= range(len(dula_data[.2]))
x=[t*100 for t in x]
```

```python
figure, axis = plt.subplots(1, 2, sharey=True, figsize=(8, 4))
dula_graph = axis[0]
for st_s in step_size:
    dula_graph.plot(x, dula_data[st_s], lw=1.5, label=f'{st_s}')
dula_graph.set_xlabel("Iteration")
dula_graph.set_ylabel("Log RMSE")
dula_graph.legend(title="Step-size")
dula_graph.set_title("DULA")

cyc_dula_graph = axis[1] 
for st_s in step_size: 
    cyc_dula_graph.plot(x, cyc_dula_data[st_s], lw=1.5, label=f'{st_s}')
cyc_dula_graph.set_xlabel("Iteration")
cyc_dula_graph.legend(title="Step-size")
cyc_dula_graph.set_title("cyclical-DULA")
plt.suptitle("Log RMSE vIteration")
```
